```python
# python 스타일 분할정복
def find_peak_in_2d_array_with_python_style(new_array):
    center = len(new_array) // 2 # 2차원 배열에서 중앙 값
    j = new_array[center].index(max(new_array[center])) # 중앙값에서 최댓값

    if len(new_array) == 1:
        return new_array[0][j]
    center = len(new_array) // 2
    if (0 < center) and (new_array[center - 1][j] > new_array[center][j]):
        return find_peak_in_2d_array_with_python_style(new_array[:center])
    elif (center < len(new_array) - 1) and (new_array[center][j] < new_array[center + 1][j]):
        return find_peak_in_2d_array_with_python_style(new_array[center:])
    else:
        return new_array[center][j]
```

```python
answer_in_python = find_peak_in_2d_array_with_python_style(array_2d)
```
